```python
import numpy as np, pandas as pd

np.random.seed(42)
n = 100000
preg = np.random.poisson(1.8, n).clip(0,17).astype(int)
age = np.random.randint(21,90,n)
glucose = np.random.normal(120,30,n).clip(50,200)
bp = np.random.normal(70,12,n).clip(40,122)
skin = np.random.normal(20,10,n).clip(0,99)
insulin = np.clip(glucose*2.5 + np.random.normal(0,80,n),0,846)
bmi = np.clip(np.random.normal(30,7,n)+0.02*(age-40)+0.01*(glucose-120)+np.random.normal(0,2,n),15,60)
ped = np.random.beta(1.5,10,n)*2.5

score = (0.0008*glucose**2 + 0.002*bmi**2 + 0.01*preg*glucose
         -0.03*bp + 0.5*np.tanh((np.log1p(insulin)-4)/2)
         +0.6*ped)
prob = 1/(1+np.exp(-(score-10)/3))
outcome = (np.random.rand(n) < prob).astype(int)

data = pd.DataFrame({
    "Pregnancies": preg, "Glucose": glucose.round(2), "BloodPressure": bp.round(2),
    "SkinThickness": skin.round(2), "Insulin": insulin.round(2), "BMI": bmi.round(2),
    "DiabetesPedigree": ped.round(4), "Age": age, "Outcome": outcome
})
data.to_csv("synthetic_diabetes_100k.csv", index=False)
print("Saved synthetic_diabetes.csv")
```
